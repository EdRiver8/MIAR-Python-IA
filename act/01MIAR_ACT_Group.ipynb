## Actividad 01

- Crear una lista con números que comiencen por 30 y que termine en 42 haciendo que el paso entre cada número consecutivo sea de 0.4. Sin usar Numpy

```python
# Your Solution Here!!!
lista = []
valor = 30
while valor <= 42:
  lista.append(round(valor, 2))
  valor += 0.4

print(lista)
```

```python
resultado = []
cantidad_de_pasos = int((42 - 30)/0.4)
for i in range(cantidad_de_pasos + 1):
  resultado.append(30 + i * 0.4)

print(resultado)
```

- Obtener el número de ciudades que tiene el estado de Florida, usando dicho Dataset
- Sin usar Pandas

```python
import urllib

url = "https://raw.githubusercontent.com/grammakov/USA-cities-and-states/master/us_cities_states_counties.csv"

# Devuelve un fichero desde una url como cadena de caracteres
def read_url(url):
    return urllib.request.urlopen(url).read().decode('utf-8')

# Descargar un fichero desde una url a una ruta local
def download_file(url, dest):
    urllib.request.urlretrieve(url, dest)
    return
```

```python
# Your Solution Here!!!
import os

dest = os.path.join("..","res", "test.csv")
download_file(url, dest)
```

```python
delimiter = "|"

def get_column_value(csv_line, idx: int):
  return csv_line.split(delimiter)[idx]

def get_city(csv_line):
  return get_column_value(csv_line, 0)

def get_state(csv_line):
  return get_column_value(csv_line, 2)

def is_state(csv_line, expected_state):
  return get_state(csv_line).upper() == expected_state.upper()


with open(dest) as csv_file:
  florida_city_set = { get_city(csv_line) for csv_line in csv_file if is_state(csv_line, "FLORIDA") }
  print(f"Número de ciudades que tiene el estado de Florida: {len(florida_city_set)}")
```

```python
import csv

set_cities = set()
with open(dest) as f:
  reader = csv.reader(f, delimiter = "|")
  for row in reader:
    state_short = row[1]
    city = row[0]
    if state_short == "FL":
        set_cities.add(city)



print(len(set_cities))
```

```python
with open(dest, "r", encoding="utf-8") as f:
  lines = f.readlines()

florida_cities = set()
lines = lines[1:]
for line in lines:
  parts = line.strip().split("|")
  city, state_short, state_full, county, alias = parts
  if state_full == "Florida":
    florida_cities.add(city)



print("Número de ciudades en Florida:", len(florida_cities))
```
