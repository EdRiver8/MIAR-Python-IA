### Renombrar y Ordenar Columnas en un Dataframe de Pandas
- Se utiliza el set de info pais

```python
info_pais = pd.read_csv(r'info_pais.csv', encoding="ISO-8859-1", delimiter=";")
info_pais
```

```python
# Renombrar columnas
info_pais.rename(columns={
    'País': 'Country',
    'Poblacion': 'Population',
    'Renta per capita': 'PIB_per_capita_USD',
    'Esperanza de vida': 'Life_expectancy'
},
    inplace=True
)
info_pais
```

```python
# Reordenar columnas
info_pais = info_pais[[
    'Country',
    'PIB_per_capita_USD',
    'Population',
    'Life_expectancy'
]]
info_pais
```

## PIVOT TABLES

- Se usan para crear tablas resumen a partir de un DataFrame, permitiendo agrupar y agregar datos de manera flexible.
- Son como una tabla dinamica en Excel.
- La idea es convertir las filas en columnas y los valores que van en ellas, usar funciones de 
agregacion como suma, promedio, conteo, etc.

```python
# Crear una tabla pivote para el set de paises
info_pais.info()
```

```python
# Como Life_expectancy es de tipo object se debe asegurar que se tome como numerico
info_pais['Life_expectancy'] = pd.to_numeric(info_pais['Life_expectancy'], errors='coerce')
info_pais.info()
```

```python
info_pais.Life_expectancy.notna().count()
```

```python
# Como hay valores nulos en la columnna de life_expectancy, se reemplazar con su mediana
info_pais['Life_expectancy'].fillna(info_pais['Life_expectancy'].median(), inplace=True)
info_pais.info()
```

```python
info_pais_pivot = info_pais.pivot_table(
    index='Country',
    values=['Population', 'PIB_per_capita_USD', 'Life_expectancy'],
    aggfunc={
        'Population': 'sum',
        'PIB_per_capita_USD': 'mean',
        'Life_expectancy': 'mean'
    }
)
info_pais_pivot
```
